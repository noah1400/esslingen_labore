```python
import pandas as pd
from scipy.stats import hypergeom
from scipy.stats import binom
TOTAL_FIELDS = 70
WINS_10_GUESSES =[
    100000*2-2,
    1000*2-2,
    100*2-2,
    15*2-2,
    5*2-2,
    2*2-2,
    -2,
    -2,
    -2,
    -2,
    2,
]
WINS_9_GUESSES =[
    50000*2-2,
    1000*2-2,
    20*2-2,
    5*2-2,
    2*2-2,
    -2,
    -2,
    -2,
    -2,
    2,
]
WINS_8_GUESSES =[
    10000*2-2,
    100*2-2,
    15*2-2,
    2*2-2,
    1*2-2,
    -2,
    -2,
    -2,
    1*2-2,
]
WINS_7_GUESSES =[
    1000*2-2,
    100*2-2,
    12*2-2,
    1*2-2,
    -2,
    -2,
    -2,
    -2,
]
WINS_6_GUESSES =[
    500*2-2,
    15*2-2,
    2*2-2,
    1*2-2,
    -2,
    -2,
    -2,
]
WINS_5_GUESSES =[
    100*2-2,
    7*2-2,
    2*2-2,
    -2,
    -2,
    -2,
]
WINS_4_GUESSES =[
    22*2-2,
    2*2-2,
    1*2-2,
    -2,
    -2,
]
WINS_3_GUESSES =[
    16*2-2,
    1*2-2,
    -2,
    -2,
]
WINS_2_GUESSES =[
    6*2-2,
    -2,
    -2,
]
```

```python
# X10
x10A = []
for i in range(11):
    x10A.append(hypergeom.pmf(i, TOTAL_FIELDS,10,20))

# Erwartungswert
erw10 = 0
for i in range(11):
    erw10 += (WINS_10_GUESSES[i]*x10A[10-i])

# Varianz
var10 = 0
for i in range(11):
    var10 += (WINS_10_GUESSES[i]**2*x10A[10-i])
var10 -= erw10**2

# Standardabweichung
std10 = var10**0.5
```
